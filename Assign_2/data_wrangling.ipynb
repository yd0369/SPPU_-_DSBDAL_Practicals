```python
from scipy.io import arff
import pandas as PD
import urllib.request
import io # for io.StringIO()



URL = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00467/Sapfile1.arff'
ftpstream = urllib.request.urlopen(URL)



# RAW_DATA = arff.loadarff('student_academic_performance_dataset.arff')
RAW_DATA = arff.loadarff(io.StringIO(ftpstream.read().decode('utf-8')))
DF = PD.DataFrame(RAW_DATA[0])

DF
```

```python
DF['ge'] = DF['ge'].str.decode('utf-8') 
DF['cst'] = DF['cst'].str.decode('utf-8') 
DF['tnp'] = DF['tnp'].str.decode('utf-8') 
DF['twp'] = DF['twp'].str.decode('utf-8') 
DF['iap'] = DF['iap'].str.decode('utf-8') 
DF['esp'] = DF['esp'].str.decode('utf-8') 
DF['arr'] = DF['arr'].str.decode('utf-8') 
DF['ms'] = DF['ms'].str.decode('utf-8') 
DF['ls'] = DF['ls'].str.decode('utf-8') 
DF['as'] = DF['as'].str.decode('utf-8') 
DF['fmi'] = DF['fmi'].str.decode('utf-8') 
DF['fs'] = DF['fs'].str.decode('utf-8') 
DF['fq'] = DF['fq'].str.decode('utf-8') 
DF['mq'] = DF['mq'].str.decode('utf-8') 
DF['fo'] = DF['fo'].str.decode('utf-8') 
DF['mo'] = DF['mo'].str.decode('utf-8') 
DF['nf'] = DF['nf'].str.decode('utf-8') 
DF['sh'] = DF['sh'].str.decode('utf-8') 
DF['ss'] = DF['ss'].str.decode('utf-8') 
DF['me'] = DF['me'].str.decode('utf-8') 
DF['tt'] = DF['tt'].str.decode('utf-8') 
DF['atd'] = DF['atd'].str.decode('utf-8') 

DF
```

### Renaming Columns

```python
DF.columns
```

```python
DF = DF.rename(columns={
    'ge'  : 'gender',
    'cst' : 'cast',
    # 'tnp' : '',
    # 'twp' : '',
    # 'iap' : '',
    # 'esp' : '',
    # 'arr' : '',
    'ms'  : 'marital_status',
    # 'ls'  : '',
    # 'as'  : '',
    # 'fmi' : '',
    # 'fs'  : '',
    # 'fq'  : '',
    # 'mq'  : '',
    'fo'  : 'father_occupation',
    'mo'  : 'mother_occupation',
    # 'nf'  : '',
    # 'sh'  : '',
    # 'ss'  : '',
    'me'  : 'mother_tongue',
    # 'tt'  : '',
    # 'atd' : ''

})

DF.columns
```
